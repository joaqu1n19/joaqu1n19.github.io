```python
import pandas as pd
titanic = pd.read_csv("https://raw.githubusercontent.com/ywen2021/CPSC392/main/Data//titanic.csv")

passenger4_and_12 = titanic.iloc[[3, 11]]
print(passenger4_and_12)
```

```python
mean_age = titanic['age'].mean()

print(f"The mean age of titanic passengers is: {mean_age:.2f}")
```

```python
penguins['bill_ratio'] = penguins['bill_length_mm'] / penguins['bill_depth_mm']

print(penguins['bill_ratio'])
```

```python
lil_boys = penguins[penguins['body_mass_g'] < 4000]

display(lil_boys.head(4))
```

```python
mean_bill_lengths = penguins.groupby('species')['bill_length_mm'].mean()

bl_Adelie = mean_bill_lengths['Adelie']
bl_Chinstrap = mean_bill_lengths['Chinstrap']
bl_Gentoo = mean_bill_lengths['Gentoo']

print(f"Mean bill length for Adelie: {bl_Adelie:.2f} mm")
print(f"Mean bill length for Chinstrap: {bl_Chinstrap:.2f} mm")
print(f"Mean bill length for Gentoo: {bl_Gentoo:.2f} mm")
```

```python
survival_proportion_by_class = titanic.groupby('class')['survived'].mean()

print("Proportion of survivors by class:")
print(survival_proportion_by_class)

highest_proportion_class = survival_proportion_by_class.idxmax()
highest_proportion_value = survival_proportion_by_class.max()

print(f"\n{highest_proportion_class} class had the highest proportion of survivors with {highest_proportion_value:.2f}")
```

```python
average_fare_by_sex = titanic.groupby('sex')['fare'].mean()

print("Average fare paid by males and females:")
print(average_fare_by_sex)

male_fare = average_fare_by_sex['male']
female_fare = average_fare_by_sex['female']

print(f"\nThe average fare for males was: ${male_fare:.2f}")
print(f"The average fare for females was: ${female_fare:.2f}")

if male_fare != female_fare:
    print(f"There is a difference of ${abs(male_fare - female_fare):.2f} in average fare paid between males and females.")
else:
    print("There is no difference in average fare paid between males and females.")
```

```python
small_penguin = penguins[['bill_length_mm', 'species', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

display(small_penguin.head())
```

```python
small_penguin = small_penguin.drop(columns=['flipper_length_mm'])

display(small_penguin.head())
```

```python
print("Info for small_penguin DataFrame:")
small_penguin.info()

print("\nDescriptive statistics for small_penguin DataFrame:")
display(small_penguin.describe())
```

```python
species_avg_bill_length = small_penguin.groupby('species')['bill_length_mm'].mean()

print("Species average bill length:")
print(species_avg_bill_length)
```

```python
filtered_penguins = small_penguin[(small_penguin['bill_length_mm'] < 39) & (small_penguin['bill_depth_mm'] < 18)]

print("Penguins with bill length < 39 and bill depth < 18:")
print(filtered_penguins)
```
